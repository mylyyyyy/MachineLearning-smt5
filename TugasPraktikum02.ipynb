## 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
used_variables = [
    "diagnosis", "radius_mean", "texture_mean", "perimeter_mean", "area_mean",
    "smoothness_mean", "compactness_mean", "concavity_mean",
    "concave points_mean", "symmetry_mean", "fractal_dimension_mean"
]

df_used = df[used_variables]

# Melihat beberapa baris pertama dari DataFrame yang sudah difilter
df_used.head()
```

## 2. Lakukan proses encoding pada kolom "diagnosis"

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder() # membuat objek dari LabelEncoder
df_used['diagnosis'] = le.fit_transform(df_used['diagnosis'])
df_used.tail(10)
```

## 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df_used['radius_mean'] = std.fit_transform(df_used[['radius_mean']])
df_used['texture_mean'] = std.fit_transform(df_used[['texture_mean']])
df_used['perimeter_mean'] = std.fit_transform(df_used[['perimeter_mean']])
df_used['area_mean'] = std.fit_transform(df_used[['area_mean']])
```

```python
df_used.head()
```

## 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Import pustaka untuk membagi data
from sklearn.model_selection import train_test_split

# Split data training dan testing
# Rasio yang akan kita gunakan adalah 8:2
df_train, df_test = train_test_split(df_used, test_size=0.2, random_state=42)

# df_train akan berisi 80% data, df_test akan berisi 20% data
# random_state digunakan untuk memastikan bahwa pembagian data adalah sama setiap kali kode dijalankan

# Mencetak ukuran dari data training dan testing
print("Ukuran Data Training:", df_train.shape)
print("Ukuran Data Testing:", df_test.shape)
```
